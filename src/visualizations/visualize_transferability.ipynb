### Visualize the model sparsity versus the model's performance and its transferability on the target dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Sparsity = np.array([100, 80, 64, 51.2,  40.97, 32.78, 26.227, 20.986, 16.79,  13.43, 10.75, 8.606, 6.889, 5.514, 4.415, 3.536, 2.832, 2.269, 1.819])
Accuracy_Sources = [83.860, 88.099, 88.389, 88.659, 88.309, 87.959, 88.610, 88.5, 88.409, 88.37, 88.18, 88.579, 88.119, 88.040, 87.459, 87.879, 87.299, 87.40, 87.229]
Accuracy_Targets = [96.399, 96.779, 96.699, 96.849, 96.379, 96.68, 96.589, 96.22, 96.449, 95.930, 95.70, 95.729, 95.860, 95.269, 94.599, 94.089, 93.459, 92.470, 91.149]
```

```python
# Visualize the sparsity from 100 to smaller values 
plt.figure(figsize=(10, 5))
plt.plot(Sparsity, Accuracy_Sources, label='Source')
plt.plot(Sparsity, Accuracy_Targets, label='Target')
# X-axis use the log scale on log8
# Reverse the x-axis
plt.xlim(101, 1.8)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Sparsity')
plt.ylabel('Accuracy')
plt.title('Transferability')
plt.legend()
plt.show()
```
