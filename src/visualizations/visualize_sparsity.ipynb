```python
import matplotlib.pyplot as plt 
import  numpy as np
```

```python
targets = ["MNIST", "SVHN", "MNIST-M", "SYN"]
model_sparsity = [0.1, 0.5,  0.9, 0.95, 0.99]

results = {}
results[targets[0]] = [0.008, 0.038, 0.001, -0.061, -0.958]
results[targets[1]] = [0.002, 0.026, -0.190, -0.336, -1.600]
results[targets[2]] = [0.051, 0.050, -0.115, -0.275, -1.623]
results[targets[3]] = [0.010, 0.057, -0.165, -0.275, -1.927]
```

```python
# plot the results as curves using tab10 color map
plt.figure(figsize=(4, 2))
markers = ['o', 's', 'd', '^', 'x']
for i, target in enumerate(targets):
    plt.plot(model_sparsity, results[target], label=target, marker=markers[i], linestyle='-', color=plt.cm.tab10(i), alpha=0.8)
plt.xscale('logit')
plt.xlabel('Model Sparsity')
plt.ylabel('SLC-AUC')
plt.xticks(model_sparsity, model_sparsity)
plt.grid()
plt.legend(title='Target Domain')
plt.tight_layout()
plt.savefig('../../imgs/sparsity_vs_slauc.pdf')
```
